# [Problem 4] Data exploration

```python
import matplotlib.pyplot as plt
```

```python
data.shape
```

```python
plt.plot(data["AMT_INCOME_TOTAL"])
plt.show()
```

```python
plt.plot(data["AMT_CREDIT"])
plt.show()
```

```python
data['diff_inc_cr'] = abs(data["AMT_INCOME_TOTAL"] - data["AMT_CREDIT"])
```

```python
plt.plot(data["diff_inc_cr"])
plt.show()
```

```python
value_counts = data['NAME_INCOME_TYPE'].value_counts()

print(value_counts)
```

```python
# Plot the counts
plt.figure(figsize=(8, 6))
value_counts.plot(kind='bar')
plt.title('Counts of Unique Values')
plt.xlabel('Unique Values')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
child_counts = data['CNT_CHILDREN'].value_counts()

print(child_counts)
```

```python
plt.figure(figsize=(8, 6))
child_counts.plot(kind='bar')
plt.title('Counts of Unique Values')
plt.xlabel('Unique Values')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
	
contract_counts = data['NAME_CONTRACT_TYPE'].value_counts()

print(contract_counts)
plt.figure(figsize=(8, 6))
contract_counts.plot(kind='bar')
plt.title('Counts of Unique Values')
plt.xlabel('Unique Values')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
