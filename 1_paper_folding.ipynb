# [Problem 1] Implementation using exponentiation arithmetic operators

```python
# Code to calculate the thickness when the paper is folded once

THICKNESS = 0.00008
folds = 43
folded_thickness = THICKNESS * (2 ** folds)

print("Thickness: {} meters".format(folded_thickness))
```

# [Problem 2] Unit Conversion

```python
print("Thickness: {: .2f} ten thousand kilometers".format(folded_thickness / (1000 * 10000)))
```

# [Problem 3] Create using a for statement

```python
for i in range(folds):
    THICKNESS = THICKNESS * 2
print("Thickness: {} meters".format(folded_thickness))
```

# [Problem 4] Comparison of calculation time

```python
import time
```

```python
# Method 1
start = time.time()
#####
folded_thickness = THICKNESS * (2 ** folds)
#####
elapsed_time = time.time() - start
print("Method 1 execution time : {}[s]".format(elapsed_time*(10**6)))
```

```python
# Method 2
start = time.time()
#####
for i in range(folds):
    THICKNESS = THICKNESS * 2
#####
elapsed_time = time.time() - start
print("Method 2 execution time : {}[s]".format(elapsed_time*(10**6)))
```

# [Problem 5] Saving to a list

```python
folded_thickness = []
THICKNESS=0.00008
folded_thickness.append(THICKNESS) # init
for i in range(folds):    
    THICKNESS = THICKNESS * 2
    folded_thickness.append(THICKNESS) # each
folded_thickness
```

# [Problem 6] Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Display the graph. Title and axis label name.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness)  # Enter the variable name of the list in "List name"
plt.show()
```

# [Problem 7] Customizing graphs

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness, color='red')
plt.show()
```
