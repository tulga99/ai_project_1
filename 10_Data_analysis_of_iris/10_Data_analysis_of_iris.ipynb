```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()

X = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])

y = pd.DataFrame(iris.target, columns=['Species'])
```

# [Problem 6] Creating a diagram

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pie chart
plt.pie(df['Species'].value_counts(), labels=iris.target_names, autopct='%1.1f%%')
plt.title('Number of Samples per Label')
plt.show()
```

```python
# Box plot
sns.boxplot(x='Species', y='sepal_length', data=df)
plt.title('Distribution of Sepal Length for Each Label')
plt.show()
```

```python
# Violin plot
sns.violinplot(x='Species', y='sepal_length', data=df)
plt.title('Distribution of Sepal Length for Each Label')
plt.show()
```

## box and violon plots main differences are box plots are easier to view outliners, violon plots are easier to view data distribution to which direction

# Problem 7] Confirming the relationship between features

```python
# Scatter plots
sns.scatterplot(x='sepal_length', y='sepal_width', hue='Species', data=df)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
# Scatterplot matrix
sns.pairplot(df, hue='Species')
plt.suptitle('Scatterplot Matrix', y=1.02)
plt.show()

# Correlation coefficient matrix
correlation_matrix = df.corr()

# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Coefficient Matrix')
plt.show()
```
