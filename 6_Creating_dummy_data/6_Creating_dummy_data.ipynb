```python
import numpy as np
```

# [Problem 1] Creating random numbers

```python
np.random.seed(0)
```

```python
mean = (-3, 0)
```

```python
covariance_matrix = [[1.0, 0.8], [0.8, 1.0]]
```

```python
random_numbers = np.random.multivariate_normal(mean, covariance_matrix, size=500)

print(random_numbers)
```

```python
random_numbers.shape
```

# [Problem 2] Visualization by scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x_coordinates = random_numbers[:, 0]
y_coordinates = random_numbers[:, 1]
```

```python
plt.scatter(x_coordinates, y_coordinates)
plt.title('graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# [Problem 3] Visualization by histogram

```python
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.hist(x_coordinates, bins=80)
plt.title('Histogram of X Coordinates')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.xlim(-6,0)

plt.subplot(2, 1, 2)
plt.hist(y_coordinates, bins=80)
plt.title('Histogram of Y Coordinates')
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.xlim(-6, 6)

plt.tight_layout()
plt.show()
```

# [Problem 4] Addition of data

```python
mean2 = (0, -3)
```

```python
random_numbers2 = np.random.multivariate_normal(mean2, covariance_matrix, size=500)

print(random_numbers2)
```

```python
combined_random_numbers = np.concatenate((random_numbers, random_numbers2))

# Extract x and y coordinates
x_coordinates = combined_random_numbers[:, 0]
y_coordinates = combined_random_numbers[:, 1]

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_coordinates, y_coordinates, color='blue', alpha=0.5)
plt.title('Scatter Plot of Random Numbers with Different Means')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
